## Understand the Figure Object - OOP Approach

```python
# DATA

a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*np.sin(x)
```

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,10)
axes1.set_ylim(0,1000)
axes1.set_xlabel('x large')
axes1.set_ylabel('y large')
axes1.set_title('largegag')

# SMALL AXES
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes2.plot(x,y)
axes2.set_xlim(0,10)
axes2.set_ylim(0,1)
axes2.set_xlabel('x large')
axes2.set_ylabel('y large')
axes2.set_title('smallgag')
```

## Figure Parameters

```python
fig = plt.figure(figsize = (4,3), dpi = 100)
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(a,b)
```

## Subplots 

```python
fig, axes = plt.subplots(nrows =2, ncols =2, dpi = 100, figsize = (8,8))
a = np.linspace(0,10,100)
b = np.sin(a)*np.cos(a)

x = np.linspace(0,10,100)
y = np.exp(x)

axes[0,0].plot(a,b)
axes[0,1].plot(x,y)

fig.suptitle('Practice Plot', fontsize = 32)
# can also set label, limit, title
fig.tight_layout()  # to fix a shitty layout
```

## Styling - Legends

```python
x = np.linspace(0,10,100)
y = np.sin(x)*np.cos(x)
fig = plt.figure(dpi = 100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label = 'x vs y')
ax.plot(x,y**2, label = 'x vs y^2')
ax.legend(loc = 0)
```

## Styling - Colors, LW, L types

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color ='pink')  # RGB HEX Code , copy and paste a hexcode
ax.plot(x,2*y, color ='#2596be')
```

```python
# line width and line styles ( --. -, -., ;, 'steps')

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color ='violet', lw = 3)
ax.plot(x,2*y, color ='blue', lw = 3, ls = '--')
ax.plot(x,3*y, color ='red', lw = 3, ls = '-.')
ax.plot(x,4*y, color ='green', lw = 3, ls = ':')
```

```python
# custom lines

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x,y, color ='violet', lw = 3)
lines[0].set_dashes([5,2,10,3,1,7])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x,y, color ='green', lw = 3, marker ='o', ls = '--')

# google matplotlib markers and line styles
```
