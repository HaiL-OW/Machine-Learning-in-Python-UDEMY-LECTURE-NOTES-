```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

```python
from sklearn.decomposition import PCA
```

```python
# help(PCA)
```

```python
pc_results = PCA(n_components=2)
```

```python
pca_model.fit(scaled_X)  # calculating eigenstuffs, cov matrix
pca_model.transform(scaled_X)  # projecting / dot product
```

```python
pc_results = pca_model.fit_transform(scaled_X)
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dictionary = load_breast_cancer()
```

```python
plt.scatter(pc_results[:,0], pc_results[:,1], c = cancer_dictionary['target']) # error
```

```python
pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_, index = ['PC1', 'PC2'], columns = df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize = (20,3))
sns.heatmap(df_comp, annot = True)
```

```python
pca_30 = PCA(n_components = 30)
pca_30.fit(scaled_X)
pca_30.explained_variance_ratio_
```

```python
explained_variance = []

for n in range(1,30):
    
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30), explained_variance)
plt.xlabel('Num of Components')
plt.ylabel('Variance Explained')
```

```python
pca_15 = PCA(n_components = 15)
pca_15.fit(scaled_X)
pca_model = PCA(n_components=15)
pca_results = pca_model.fit_transform(scaled_X)
np.sum(pca_15.explained_variance_ratio_)
```
